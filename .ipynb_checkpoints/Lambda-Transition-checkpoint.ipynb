# Two-Photon Stimulated Raman Transition
In the [Richerme Quantum Simulation](https://iontrap.physics.indiana.edu) lab, we work with hyperfine qubits in $^{171}\text{Yb}^+$. To make the transition from $|0\rangle$ to $|1\rangle$, we use a two-photon stimulated raman transition at 355 nm. First, let's consider the simpler case of Rabi Flopping in a two-level atom. Then, we'll consider the more complication case of the two-photon Raman transition sometimes called a $\lambda$ transition due to its shape in an energy level diagram.

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Two-level Rabi Flopping
The math turns out as follows. (More on this later).
$$
|\psi(t)\rangle = C_a(t)|a\rangle + C_b(t)|b\rangle + C_e(t)|e\rangle
$$

Where the time dependance of the coefficients are as follows:
$$
\begin{align*}
i\dot{\widetilde{C}_a}(t) &= \frac{g}{2}\widetilde{C}_e(t)\left[e^{-i\Delta t}+e^{-i\left(\Delta+\delta\omega\right)t}\right]
\\
i\dot{\widetilde{C}_b}(t) &= \frac{g}{2}\widetilde{C}_e(t)\left[e^{-i\left(\Delta-\omega_{ba}\right)t}+e^{-i\left(\Delta-\omega_{ba}+\delta\omega\right)t}\right]
\\
i\dot{\widetilde{C}_e}(t) &= \frac{g*}{2}\left[\widetilde{C}_a(t)\left[e^{i\Delta t}+e^{i\left(\Delta+\delta\omega\right)t}\right]+\widetilde{C}_b(t)\left[e^{i\left(\Delta-\omega_{ba}\right)t}+e^{i\left(\Delta-\omega_{ba}+\delta\omega\right)t}\right]\right]
\end{align*}
$$
and
$$
C_n(t) = \widetilde{C}_n(t)e^{-i\omega_n} \text{(n = a,b,e)}
$$


Let's first solving, set $\omega_{ba}$ and $\delta\omega$ to zero. $\widetilde{C}_a(0) = 1$, $\widetilde{C}_b(0) = 0$, and $\widetilde{C}_e(0) = 0$ will be the initial conditions

```python
# Set numerical hyperparameters
n = 50000  # number of points
ts = 0  # start time
te = 0.05 # end time
t = np.linspace(ts,te,n)  # time points
tau = (te-ts)/n  # smallest time increment

# Set physical quantities
g = 4  # rabi oscillation strength
Delta = 20
wa = 15*Delta
wb = 10*Delta

# Initialize Ca, Cb, and Ce
ca = np.zeros(n,dtype=np.complex_)
cb = np.zeros(n,dtype=np.complex_)
ce = np.zeros(n,dtype=np.complex_)

# Set initial conditions
ca[0] = 1
cb[0] = 0  # redundant but makes for better reading
ce[0] = 0  # redundant but makes for better reading
```

## Discretize Derivatives

We can step through the discretized values of $C_a$, $C_b$ and $C_e$  (remebering $\omega_{ab} = \delta\omega = 0$ and $g = 1$) with the simple forward difference derivatve $f'(x)\approx\frac{f(x+h) - f(x)}{h}$. I'll also convert back to $\widetilde{C}_i \implies Ci$ for simplicity

$$
\begin{align*}
i\frac{Ca_{n+1} - Ca_{n}}{\tau} &= Ce_n e^{-i\Delta t_n}
\\
i\frac{Cb_{n+1} - Cb_{n}}{\tau} &= Ce_n e^{-i\Delta t_n}
\\
i\frac{Ce_{n+1} - Ce_{n}}{\tau} &= \left[Ca_n e^{i\left(\Delta\right)t_n} + Cb_n e^{i\left(\Delta\right)t_n}\right]
\end{align*}
$$

where $\tau$ is a smallest discrete time step. Solving for the new iteration, we can see how the eigenvalues evolve in time

$$
\begin{align*}
Ca_{n+1} &= -i\tau Ce_n e^{-i\Delta t_n} + Ca_{n}
\\
Cb_{n+1} &= -i\tau Ce_n e^{-i\Delta t_n} + Cb_{n}
\\
Ce_{n+1} &= -i\tau\left[Ca_n e^{i\left(\Delta + \omega_a\right)t_n} + Cb_n e^{i\left(\Delta +\omega_b\right)t_n}\right] + Ce_{n}
\end{align*}
$$



```python
for m in range(0,len(ca)-1):
    ca[m+1] = -1.0j*tau*ce[m] *np.exp(-1.0j*Delta*t[m]) + ca[m]
    cb[m+1] = -1.0j*tau*ce[m] *np.exp(-1.0j*Delta*t[m]) + cb[m]
    ce[m+1] = -1.0j*tau*(ca[m]*np.exp(1.0j*(Delta + wa)*t[m]) + cb[m]*np.exp(1.0j*(Delta + wb)*t[m])) + ce[m]
```

```python
pa = np.absolute(ca*ca)
pb = np.absolute(cb*cb)
pe = np.absolute(ce*ce)

plt.plot(t,pa,t,pb,t,pe)
plt.legend(['P(a)','P(b)','P(e)'])
plt.show()
```

```python
# Check if probability is conserved (should be a flat line a y = 1.0)
plt.plot(t,pa+pb+pe)
plt.show()
```
